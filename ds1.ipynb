```python
import plotly.graph_objects as go
import pandas as pd
import plotly.figure_factory as ff
from IPython.display import display, HTML

df = pd.read_csv('ds_salaries.csv')

def obj_var_summary(df, obj_var):
    avg_salary = df.groupby(obj_var)['salary_in_usd'].mean().reset_index()
    avg_salary_sorted = avg_salary.sort_values('salary_in_usd', ascending=False)
    fig = go.Figure()
    fig.add_trace(go.Bar(x=avg_salary_sorted[obj_var], y=avg_salary_sorted['salary_in_usd']))

    fig.update_layout(title=obj_var.capitalize() + ' vs Average Salary',
                      xaxis_title=obj_var.capitalize(),
                      yaxis_title='Average Salary in USD',
                      template='plotly_dark')

    fig.show()
    table_html = avg_salary_sorted.to_html(index=False, classes='dark-table')
    display(HTML(table_html))

obj_var_summary(df, 'job_title')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected = True)
import warnings
warnings.filterwarnings("ignore")

df=pd.read_csv('ds_salaries.csv')

def df_corr(df):
    sns.set_theme(style="dark")
    plt.figure(figsize = (10,5.5))
    corr = df.corr()
    matrix = np.triu(corr)
    sns.heatmap(corr, annot = True, mask = matrix, cmap = "gist_gray")
df_corr(df)
```

```python
import pandas as pd
import plotly.express as px
import plotly.figure_factory as ff
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected = True)

df=pd.read_csv('ds_salaries.csv')

def display_topn_cat_val(df, feature, n=5):
    topn=df[feature].value_counts()[:n]
    fig=px.bar(y=topn.values, x=topn.index,
              text=topn.values, title= 'Top {} Values'.format(n))
    fig.update_layout(title={'text': feature,
                            'y':0.9,
                            'x':0.5,
                            'xanchor':'center',
                            'yanchor': 'top'},
                     barmode='overlay',
                     yaxis=dict(title='count'),
                     template='plotly_dark')
    fig.show()
    
display_topn_cat_val(df, 'job_title', n=10)
    
```
